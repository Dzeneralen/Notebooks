# Basics of matplotlib with numpy arrays

````python
# import both libraries
import numpy as np
import matplotlib.pyplot as plt
````

### Function documentation


## Line Plots

````python
# create a line_indices array with values from 0 - 2pi split in 50 parts
line_indices = np.linspace(0, 2 * np.pi, 50)
len(line_indices)
````

````python
# create a basic sinus plot 
plt.plot(line_indices, np.sin(line_indices))
plt.show()
````

### Additional styling

````python
# red solid line with crosses at the indices
plt.plot(line_indices, np.sin(line_indices), 'r-+')
plt.show()
````

### Adding a legend, title and axis labels

````python
# add a title to the plot
plt.title('Hello world of plots!')

# add a label on the x-axis
plt.xlabel('This is the x-axis')

# add a label on the y-axis
plt.ylabel('This is the y-axis')

# plot of sin and cos with respective legends by using the label param
plt.plot(line_indices, np.sin(line_indices), label='sin')
plt.plot(line_indices, np.cos(line_indices), label='cos')

# call legend to show the legend
plt.legend()

plt.show()
````

## Scatter plots

````python
# use the same indices for the line plot, but show a scatterplot instead
plt.scatter(line_indices, np.sin(line_indices))
plt.show()
````

````python
# you can achieve the same thing with lineplots, just remove the line styling and add the dot at indices
plt.plot(line_indices, np.sin(line_indices), 'oc')
plt.show()
````

### More advanced scatter plots
You can create some decent looking scatterplots by applying weight and size to the indices.

````python
random_x_arr = np.random.rand(100)
random_y_arr = np.random.rand(100)
random_size_arr = np.random.rand(100) * 50
random_color_arr = np.random.rand(100)

plt.scatter(random_x_arr, random_y_arr, random_size_arr, random_color_arr)
plt.colorbar()
plt.show()
````

## Multiple plots on the same figure
To achieve multiple plots on the same figure you can use subplot. Subplot takes inn the number of rows, columns and which plot to activate. **N.B. subplot uses 1-indexing instead of the default 0-indexing in Python.**

````python
# activate the first subplot
plt.subplot(2,2,1) 
plt.plot(line_indices, np.sin(line_indices), 'b-.')

# activate the second subplot
plt.subplot(2,2,2)
plt.plot(line_indices, np.cos(line_indices), 'r--')

# activate the second subplot
plt.subplot(2,2,3) 
plt.plot(line_indices, np.sin(line_indices), 'b-.')
plt.plot(line_indices, np.cos(line_indices), 'r--')

plt.subplot(2,2,4) 

plt.show()
````

### Useful function when working with many subplots
When working with multiple subplots we can use the 
```python 
figure, axes = plt.subplots(num_rows,num_cols, sharex=True, sharey=True)
``` 
to get back a figure, and an array containing axes for our plots **0-indexed by [row][col].**

When using axes we can set the title, and legends for the axis by using `set_title`, `set_xlabel` and `set_ylabel`.

````python
figure, axes = plt.subplots(2, 2, sharex=True, sharey=True)

# plot a blue graph in the upper left corner
axes[0][0].plot(line_indices, np.sin(line_indices), 'b--')
axes[0][0].set_title('[0][0]')

# plot a red graph in the upper right corner
axes[0][1].plot(line_indices, np.cos(line_indices), 'r')
axes[0][1].set_title('[0][1]')

# plot a green graph in the lower left corner
axes[1, 0].plot(line_indices, np.cos(line_indices), 'g')
axes[1][0].set_title('[1][0]')

# plot a magenta graph in the lower right corner
axes[1, 1].plot(line_indices, np.sin(line_indices), 'm--')
axes[1][1].set_title('[1][1]')

# we need to call plt.show() here and not figure.show(), plt.show() will draw all the figures for us in non-interactive mode
plt.show()
````

## Histograms
To plot histograms we can use the hist method on matplotlib. Here we show its basic operations, but plenty options are available such as normalizing the data before plotting.

````python
# basic histogram of 1000 random numbers between 0 - 1 sampled from a normal distribution (randn, not rand)
plt.hist(np.random.randn(1000))
plt.show()
````

````python
# basic histogram of 1000 random numbers between 0 - 1 sampled from a uniform distribution (rand) with 30 bins
plt.hist(np.random.rand(1000), 30, histtype='step', color='m')
plt.show()
````

## Error plots

````python
# basic error plot with the x-values, y-values and the y-errors. (x-errors can also be passed as the fourth parameter)
plt.errorbar(line_indices, np.sin(line_indices), np.random.rand(len(line_indices)) / 5, color='b', ecolor='orange')
plt.show()
````
